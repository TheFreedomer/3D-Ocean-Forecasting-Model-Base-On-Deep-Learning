```python
import numpy as np
import matplotlib.pyplot as plt

# 生成示例数据（替换为你的数组）
data = np.random.randn(1000)  # 正态分布的1000个随机数

# 自定义区间边界
bins = np.linspace(-3, 3, 13)  # 从-3到3划分12个区间

plt.hist(data, bins=bins,
         edgecolor='white',
         color='skyblue',
         alpha=0.7)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
