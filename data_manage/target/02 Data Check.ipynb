```python
import os
import netCDF4 as nc
import numpy as np
import pandas as pd
```

```python
dir_base = r'E:\Dataset_SCS\ERA5'
filenames = os.listdir(dir_base)
```

```python
def check_nc_integrity_basic(file_path):
    """基本完整性检查"""
    try:
        with nc.Dataset(file_path, 'r') as ds:
            print(f"文件 '{file_path}' 可以正常打开")
            print(f"文件格式: {ds.data_model}")
            print(f"变量数量: {len(ds.variables)}")
            print(f"维度数量: {len(ds.dimensions)}")
            return True
    except Exception as e:
        print(f"文件 '{file_path}' 损坏或无法读取: {str(e)}")
        return False
```

```python
lst_bad = []
for filename in filenames:
    filepath_temp = os.path.join(dir_base, filename)
    condition = check_nc_integrity_basic(filepath_temp)
    if not condition:
        lst_bad.append(filename)
```
