```python
import os
import numpy as np
import pickle
```

```python
dir_base = r'E:\dataset\dataset'
```

```python
idx_s, idx_e = 0, 10
lst_variable = ['tx', 'ty', 'ssr', 'str', 'ssh', 'u', 'v', 't', 's']
is_2d = [True, True, True, True, True, False, False, False, False]
is_mask = [False, False, False, False, True, True, True, True, True]
```

```python
def calculate_measure(file_names, file_dir_data, file_dir_mask=None, show=False):
    """

    :param file_names:
    :param file_dir_data:
    :param file_dir_mask:
    :param show:
    :return:
    """
    # mean
    _sum = 0
    _n = 0
    for file_name in file_names:
        if show:
            print(file_name)
        filepath_data_temp = os.path.join(file_dir_data, file_name)
        data_temp = np.load(filepath_data_temp)
        if file_dir_mask:
            filepath_mask_temp = os.path.join(file_dir_mask, file_name)
            mask_temp = np.load(filepath_mask_temp)
            data_temp[mask_temp] = np.nan
        _sum += np.nanmean(data_temp)
        _n += 1
    _mean = _sum / _n

    # std
    _sum = 0
    _n = 0
    for file_name in file_names:
        if show:
            print(file_name)
        filepath_data_temp = os.path.join(file_dir_data, file_name)
        data_temp = np.load(filepath_data_temp)
        if file_dir_mask:
            filepath_mask_temp = os.path.join(file_dir_mask, file_name)
            mask_temp = np.load(filepath_mask_temp)
            data_temp[mask_temp] = np.nan
        _sum += np.nanmean((data_temp - _mean) ** 2)
        _n += 1
    _std = np.sqrt(_sum / _n)

    return _mean, _std
```

```python
dict_statistics = {}
for v, variable in enumerate(lst_variable):
    print('variable:', variable)
    filenames = sorted(os.listdir(os.path.join(dir_base, 'data', variable)))
    dict_statistics[variable] = {}
    # print(filenames)
    if is_2d[v]:
        print('is_2d')
        dir_sub_data = os.path.join(dir_base, 'data', variable)
        filenames_final = filenames
        if is_mask[v]:
            print('is_mask')
            dir_sub_mask = os.path.join(dir_base, 'mask', variable)
        else:
            dir_sub_mask = None
        mean_temp, std_temp = calculate_measure(filenames_final, dir_sub_data, dir_sub_mask, show=False)
        dict_statistics[variable]['mean'] = mean_temp
        dict_statistics[variable]['std'] = std_temp
        print(mean_temp, std_temp)
    else:
        for folder_level in filenames:
            print(folder_level)
            dict_statistics[variable][folder_level] = {}
            dir_sub_data = os.path.join(dir_base, 'data', variable, folder_level)
            filenames_final = sorted(os.listdir(dir_sub_data))
            if is_mask[v]:
                dir_sub_mask = os.path.join(dir_base, 'mask', variable, folder_level)
            else:
                dir_sub_mask = None
            mean_temp, std_temp = calculate_measure(filenames_final, dir_sub_data, dir_sub_mask, show=False)
            dict_statistics[variable][folder_level]['mean'] = mean_temp
            dict_statistics[variable][folder_level]['std'] = std_temp
            print(mean_temp, std_temp)

# save
with open('../../data/statistics/all/statistics.pkl', 'wb') as f:
    pickle.dump(dict_statistics, f)
```
